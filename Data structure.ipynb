5. Why might you use a set instead of a list in Python ?

 ▶ Sets and lists sevre different purposes in python, leading to situations where one is more suitable than the other. Sets are particularly useful when:

Uniqueness is required:-
Sets,by defination,only store unique elements. if the goal is to eliminate duplicates from a collection, converting it to set is an efficient approach.

Membership testing is frequent:-
Sets are optimized for fast membership testing (checking if element is present). This operation has an average time complexity of O(1) for sets, compared to O(n) is the number of elements. When performing many membership checks, using a set can lead to significant performance improvements.

Set operations are needed:- Sets support common mathematical set operations like union, intersection and difference. If these operations are necessary, using sets directly is more convient and efficient than implementing them manually with lists.

 ▶ In contrast, lists are more appropriate when:

Order matters:-
Lists maintain the order of elements, while sets do not gurantee any specific order. If the sequence of elements is important, a list should be used.

Duplicates are allowed:-
If duplicate elements are acceptable or required, a list is the natural choice.

Indexing or slicing is neccesary:-
Lists support indexing and slicing, allowing access to elements at specific order. Sets do not provide this functionality.

 ▶ For example, consider a scenario where the task is to find the unique words in a large corpus. Using a set would be efficient for this task, as it automatically handles duplicate elimination and provide fast membership testing for checking if a word has already been encountered.

8. What is a hash table and how does it relate to dictionaries in Python ?

 ▶ A hash table is a data structure that stores key-value pairs, where each key is unique and mapped to a specific location (index) in an array using a hash function. This allows for efficient retrieval,insertion and deletion of data.

 ▶Python dictionaries are implemented using hash tables. when a key-value pair is added to a dictionary, python calculates the hash value of the key and users it to determine the index where the value will be stored. When retrieving a value, python calculates the hash value of the key again and uses it to quickly locate the value in the hash table.

 ▶ Hash tables provide average-case o(1) time complexity for search, insert and delete operations, making them a hghly efficient data structure for implementing dictinaries. However, in the worst-case scenario(e.g many keys hashing to the same index), these operations can take O(n) time, where n is the number of key-value pairs. to mitigate this, python's dictionary implementation includes collision resolution mechanisms to handle mutltiple keys mapping keys mapping to same index.

10. Explain why strings are immutable in Python.

 ▶ In python, strings are immutable because of several key benefits related to efficiency, safety and simplicity.

* Efficiency:- Immutabilty allows python to optimize memory usage. when multiple string variables have the same value, they can reference the same memory location. If strings were mutable, each variable would need its own copy, wasting memory.

* Safety:- Immutability prevents unintended modifications. If a function receives a string as input, it can be sure that the original string will not be altered, avoiding unexpected side effects.This is crucial in multithread environments where data consistency is paramount

* Hashing:-Immutable objects, like strings, can be used as keys in dictinaries and elements ins sets. Mutable objects cannot be hashed because their values can change, which would violate the integrity of these data structures.

* Simplicity:- Immutability makes string manipulation more predictable. When a string is modified, a new string object is created, leaving the original string unchanged. This simplifies debugging and reasoning about code.
